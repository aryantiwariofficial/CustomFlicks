```python
from datetime import datetime
```

```python
def date_to_numerical(date_str):
    date_obj = datetime.strptime(date_str, '%d-%m-%y')
    year_start = datetime(date_obj.year, 1, 1)
    days_elapsed = (date_obj - year_start).days + 1  # Adding 1 to start counting from 1
    return days_elapsed
```

```python
for index, row in dataset.iterrows():
    date_str = row['Last Payment Date']  
    numerical_value = date_to_numerical(date_str)
    dataset.at[index, 'NumericalDate'] = numerical_value
```

```python
def assign_label(numerical_date):
    if 160 <= numerical_date <= 170:
        return 'least_interested'
    elif 171 <= numerical_date <= 180:
        return 'moderate_nterested'
    elif 181 <= numerical_date <= 200:
        return 'highly_interested'
    else:
        return 'Other'  


df['Label'] = df['NumericalDate'].apply(assign_label)


#df.to_csv('new.csv', index=False)
```
